## Line Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]

x_1 = pd.DataFrame({"var": var, "var_1": var_1})
sns.lineplot(x="var", y="var_1", data=x_1)

plt.show()
```

```python
plt.plot(var, var_1)
plt.show()
```

```python
var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]

sns.lineplot(x=var, y=var_1)

plt.show()
```

## Histogram Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

- bins argument is used to set the range of the data or organize the data
- kde(kernel density estimator) used to create line graph above this bar graph
- rug is used to show kernel density on x axis
- color is used to change the color of the bar
- log scale is used to convert the data and the graph into log

```python
sns.displot(var['flipper_length_mm'], bins=[170,180,190,200,210,220,230,240], kde=True, rug=True, color="g")  # log_scale=True
plt.show()
```

```python
sns.displot(var['flipper_length_mm'], kde=True)
plt.show()
```

## Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.set(style="darkgrid")
order_1 = ['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x='island', y='bill_length_mm', data=var, hue='sex', order=order_1, hue_order=['Female','Male'], 
            saturation=100, err_kws={'color': 'b'}, dodge=True, alpha=0.9) # ci=100, n_boot=2, palette="icefire", errcolor='y', errwidth=12, capsize=0.5
plt.show()
```

```python
sns.barplot(x='island', y='bill_length_mm', data=var, color='b')
plt.show()
```

```python
order_1 = ['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x='bill_depth_mm', y='bill_length_mm', data=var,orient='h') # ci=100, n_boot=2
plt.show()
```

# HeatMap

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var = np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var, vmin=0, vmax=10, cmap="PuOr", annot=True)
```

```python
data = sns.load_dataset("anagrams")
x = data.drop(columns=['attnr'],axis=1).head(10)
x
```

```python
sns.heatmap(x)
```

```python
var1 = np.linspace(1,10,10).reshape(2,5)
var1
```

```python
y = {"fontsize":20, "color": "r"}
v = sns.heatmap(var1, vmin=0, vmax=10, cmap="PuOr", annot=True, annot_kws=y, linewidths=10, linecolor="y", cbar=False,
            xticklabels=False, yticklabels=False)
v.set(xlabel="python", ylabel="wscube")
sns.set(font_scale=1)
plt.show()
```

```python
ar = np.array([["a0","a1","a2","a3","a4"], ["b0","b1","b2","b3","b4"]])
ar
```

```python
sns.heatmap(var1, vmin=0, vmax=10, cmap="PuOr", annot=ar, fmt="s")
```

# Box Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset('tips')
var
```

```python
sns.set_theme(style='whitegrid')
sns.boxplot(x='day', y='total_bill', data=var, hue='day',palette='plasma',order=['Fri','Sun','Thur','Sat'], showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"b"}, linewidth=3,) # palette='light:w'
plt.show()
```

```python
sns.set_theme(style='whitegrid')
sns.boxplot(x='total_bill',y='day', data=var, hue='day',palette='plasma',order=['Fri','Sun','Thur','Sat'], showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"b"}, linewidth=3,) # palette='light:w'
plt.show()
```

```python
sns.set_theme(style='whitegrid')
sns.boxplot(data=var, palette='plasma', showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"b"}, linewidth=3,orient="h") # palette='light:w'
plt.show()
```

```python
sns.set_theme(style='whitegrid')
sns.boxplot(y=var['total_bill'])
plt.show()
```

# Styling Plots in seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset('tips')
var
```

```python
sns.set_style('darkgrid')
sns.barplot(x='day', y='total_bill', data=var)
plt.show()
```

```python
sns.set_style('white')
plt.figure(figsize=(12,5))
sns.barplot(x='day', y='total_bill', data=var)
sns.despine()
plt.show()
```

```python
sns.set_style('white')
sns.set_context('paper',font_scale=1)
sns.barplot(x='day', y='total_bill', hue='day',data=var, palette='cool')
plt.show()
```
